```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('cumulative.csv')
df.head()
```

```python
df.columns
```

```python
df = df[df['koi_disposition'].isin(['CONFIRMED', 'CANDIDATE'])]

```

```python
drop_cols = [
    'rowid', 'kepid', 'kepoi_name', 'kepler_name',
    'koi_disposition', 'koi_pdisposition', 'koi_score',
    'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec',
    'koi_prad_err1', 'koi_prad_err2',
    'koi_srad_err1', 'koi_srad_err2',
    'koi_steff_err1', 'koi_steff_err2',
    'koi_slogg_err1', 'koi_slogg_err2'
]
df.drop(columns=drop_cols, inplace=True, errors='ignore')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Habitable'] = (
    (df['koi_prad'] >= 0.5) & (df['koi_prad'] <= 2.5) &
    (df['koi_teq'] >= 200) & (df['koi_teq'] <= 350)
).astype(int)
```

```python
features = ['koi_prad', 'koi_teq', 'koi_period', 'koi_steff', 'koi_srad', 'koi_slogg']
df = df[features + ['Habitable']].dropna()
```

```python
X = df[features]
y = df['Habitable']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
leak_cols = ['koi_disposition', 'koi_pdisposition', 'koi_score']
df.drop(columns=leak_cols, inplace=True, errors='ignore')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
f1_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='f1')
print("F1 scores from 5-fold CV:", f1_scores)
print("Average F1:", f1_scores.mean())
```
